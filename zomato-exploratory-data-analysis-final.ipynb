```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpfulpackages to load

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

Reading CSV

```python
df=pd.read_csv('/kaggle/input/zomato-bangalore-restaurants/zomato.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df=df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.head()
df.shape
```

Cleaning rate value

```python
df['rate'].unique()
```

Remove String and '/5'


```python
def managerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value=str(value).split('/')
        value=value[0]
        return float(value)
        
df['rate']=df['rate'].apply(managerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

Fill the null values with the mean of rate column


```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
df.rate.isnull().sum()
```

```python
df.info()
```

Deopping NULL values

```python
df.dropna(inplace=True)
df.head()
```

```python
df.info()
```

Rename columns

```python
df.rename(columns={'approx_cost(for two people)':'Costof2','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

Remove listed_in(city) and only keep location


```python
df=df.drop( ['listed_in(city)'],axis=1)
df
```

```python
df.head()
```

```python
df['Costof2'].unique()
```

Cleaning 'Costof2'

```python
def Removecomma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        
        return float(value)
df['Costof2']=df['Costof2'].apply(Removecomma)
df['Costof2'].unique()
```

Cleaning rest_type

```python
df['rest_type'].value_counts()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_typeslessthan1000=rest_types[rest_types<1000]
rest_typeslessthan1000
```

Making Rest types less than 1000 frequency as others

```python
def resttypehandle(value):
    if value in(rest_typeslessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(resttypehandle)
df['rest_type'].value_counts()
```

Cleaning Location columns

```python
df['location'].value_counts()
```

```python
Location1=df['location'].value_counts(ascending=False)
Location1
```

```python
Locationlessthan300=Location1[Location1<300]
Locationlessthan300
```

```python
def Locationhandle(value):
    if value in(Locationlessthan300):
        return 'others'
    else:
        return value
df['location']=df['location'].apply(Locationhandle)
df['location'].value_counts()
```

```python
df.head()
```

Cleaning Cuisine Column

```python
df['cuisines']
```

```python
df['cuisines'].value_counts()
```

```python
cuisines=df['cuisines'].value_counts(ascending=False)
cuisines
```

```python
Cuisinelessthan100=cuisines[cuisines<100]
Cuisinelessthan100
```

```python
def handleCuisine(value):
    if value in (Cuisinelessthan100):
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(handleCuisine)
df['cuisines'].value_counts()
```
